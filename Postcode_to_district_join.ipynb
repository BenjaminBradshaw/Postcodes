```python
import geopandas as gp
import plotly.express as px
```

```python
postcode= gp.read_file("https://www.abs.gov.au/statistics/standards/australian-statistical-geography-standard-asgs-edition-3/jul2021-jun2026/access-and-downloads/digital-boundary-files/POA_2021_AUST_GDA2020_SHP.zip")
```

```python
postcode.plot()
```

```python
postcode
```

```python
Rd =gp.read_file("https://raw.githubusercontent.com/BenjaminBradshaw/Postcodes/main/RotaryDistricts.geojson").to_crs("7844")
```

```python
Rd.plot()
```

```python
Rd
```

```python

join= Rd.sjoin(postcode,how='left',predicate='contains' )
```

```python
join
```

```python
def WhichRD(postcode):
  try:
    return join[join.POA_CODE21==postcode].RotaryDistrict.values.tolist()
  except:
    return "UNKNOW"
```

```python
WhichRD("7119")
```

```python
fig = px.choropleth(join, geojson=join.geometry, color = "POA_CODE21", projection="mercator")
fig.update_geos(fitbounds="locations", visible=True)
fig.update_layout(margin={"r":0,"t":0,"l":0,"b":0})
fig.show()                   
```

```python
fig = px.choropleth(join, ,locations="RotaryDistrict", featureidkey="properties.RotaryDistrict",projection="mercator")
fig.update_geos(fitbounds="locations", visible=True)
fig.update_layout(margin={"r":0,"t":0,"l":0,"b":0})
fig.show()    
```
